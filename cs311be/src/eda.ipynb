````python
# T·∫°o DataFrame t·ªïng h·ª£p t·ª´ t·∫•t c·∫£ c√°c file
def load_all_data():
    """Load v√† t·ªïng h·ª£p d·ªØ li·ªáu t·ª´ t·∫•t c·∫£ c√°c file CSV"""
    all_data = []
    
    for file_path in csv_files:
        try:
            # ƒê·ªçc file CSV
            df = pd.read_csv(file_path, encoding='utf-8')
            
            # Chu·∫©n h√≥a t√™n c·ªôt (m·ªôt s·ªë file c√≥ t√™n c·ªôt kh√°c nhau)
            if 'C√¢u h·ªèi' in df.columns:
                question_col = 'C√¢u h·ªèi'
                answer_col = 'Tr·∫£ l·ªùi'
            elif 'Question' in df.columns:
                question_col = 'Question'
                answer_col = 'Answer'
            else:
                continue
            
            # L·∫•y t√™n file (kh√¥ng c√≥ extension) l√†m ch·ªß ƒë·ªÅ
            topic = os.path.splitext(os.path.basename(file_path))[0]
            
            # Th√™m th√¥ng tin ch·ªß ƒë·ªÅ v√† ngu·ªìn
            df['Topic'] = topic
            df['File_Source'] = os.path.basename(file_path)
            df['Category'] = categorize_file(topic)
            
            # Chu·∫©n h√≥a t√™n c·ªôt
            df = df.rename(columns={question_col: 'Question', answer_col: 'Answer'})
            
            # Ch·ªâ gi·ªØ l·∫°i c√°c c·ªôt c·∫ßn thi·∫øt
            df = df[['Question', 'Answer', 'Topic', 'File_Source', 'Category']].copy()
            
            # Lo·∫°i b·ªè c√°c d√≤ng c√≥ d·ªØ li·ªáu thi·∫øu
            df = df.dropna(subset=['Question', 'Answer'])
            
            all_data.append(df)
            
        except Exception as e:
            print(f"‚ö†Ô∏è L·ªói khi ƒë·ªçc file {file_path}: {e}")
            continue
    
    # K·∫øt h·ª£p t·∫•t c·∫£ d·ªØ li·ªáu
    if all_data:
        combined_df = pd.concat(all_data, ignore_index=True)
        return combined_df
    else:
        return pd.DataFrame()

# Load d·ªØ li·ªáu
print("üîÑ ƒêang load d·ªØ li·ªáu t·ª´ c√°c file CSV...")
df = load_all_data()

print(f"‚úÖ ƒê√£ load th√†nh c√¥ng d·ªØ li·ªáu!")
print(f"üìä T·ªïng s·ªë d√≤ng d·ªØ li·ªáu: {len(df):,}")
print(f"üìÅ S·ªë l∆∞·ª£ng ch·ªß ƒë·ªÅ: {df['Topic'].nunique()}")
print(f"üè∑Ô∏è S·ªë l∆∞·ª£ng categories: {df['Category'].nunique()}")

# Hi·ªÉn th·ªã th√¥ng tin c∆° b·∫£n v·ªÅ DataFrame
print(f"\nüìã Th√¥ng tin c∆° b·∫£n v·ªÅ d·ªØ li·ªáu:")
print(f"K√≠ch th∆∞·ªõc: {df.shape}")
print(f"C√°c c·ªôt: {list(df.columns)}")
print(f"Ki·ªÉu d·ªØ li·ªáu:")
print(df.dtypes)
````

````python
# Hi·ªÉn th·ªã m·ªôt s·ªë m·∫´u d·ªØ li·ªáu
print("üîç M·ªôt s·ªë m·∫´u d·ªØ li·ªáu:")
print("=" * 80)
for i in range(min(3, len(df))):
    print(f"\nüìù M·∫´u {i+1}:")
    print(f"Ch·ªß ƒë·ªÅ: {df.iloc[i]['Topic']}")
    print(f"Category: {df.iloc[i]['Category']}")
    print(f"C√¢u h·ªèi: {df.iloc[i]['Question'][:100]}...")
    print(f"C√¢u tr·∫£ l·ªùi: {df.iloc[i]['Answer'][:150]}...")
    print("-" * 80)

# Ki·ªÉm tra d·ªØ li·ªáu thi·∫øu
print(f"\nüîç Ki·ªÉm tra d·ªØ li·ªáu thi·∫øu:")
missing_data = df.isnull().sum()
print(missing_data)

# Ki·ªÉm tra d·ªØ li·ªáu tr√πng l·∫∑p
duplicates = df.duplicated(subset=['Question']).sum()
print(f"\nüîÑ S·ªë c√¢u h·ªèi tr√πng l·∫∑p: {duplicates}")

# Th·ªëng k√™ theo category
print(f"\nüìä Th·ªëng k√™ theo category:")
category_stats = df['Category'].value_counts()
print(category_stats)
````

````python
# T√≠nh to√°n ƒë·ªô d√†i c√¢u h·ªèi v√† c√¢u tr·∫£ l·ªùi
df['Question_Length'] = df['Question'].str.len()
df['Answer_Length'] = df['Answer'].str.len()
df['Question_Word_Count'] = df['Question'].str.split().str.len()
df['Answer_Word_Count'] = df['Answer'].str.split().str.len()

# Th·ªëng k√™ m√¥ t·∫£ v·ªÅ ƒë·ªô d√†i
print("üìä TH·ªêNG K√ä M√î T·∫¢ V·ªÄ ƒê·ªò D√ÄI")
print("=" * 50)

# Th·ªëng k√™ c√¢u h·ªèi
print("\nüîç C√¢u h·ªèi:")
question_stats = df['Question_Length'].describe()
print(question_stats)

print(f"\nüìù S·ªë t·ª´ trung b√¨nh trong c√¢u h·ªèi: {df['Question_Word_Count'].mean():.1f}")
print(f"üìù S·ªë t·ª´ t·ªëi ƒëa trong c√¢u h·ªèi: {df['Question_Word_Count'].max()}")

# Th·ªëng k√™ c√¢u tr·∫£ l·ªùi
print("\nüí¨ C√¢u tr·∫£ l·ªùi:")
answer_stats = df['Answer_Length'].describe()
print(answer_stats)

print(f"\nüìù S·ªë t·ª´ trung b√¨nh trong c√¢u tr·∫£ l·ªùi: {df['Answer_Word_Count'].mean():.1f}")
print(f"üìù S·ªë t·ª´ t·ªëi ƒëa trong c√¢u tr·∫£ l·ªùi: {df['Answer_Word_Count'].max()}")

# Ph√¢n t√≠ch theo category
print(f"\nüìä Th·ªëng k√™ theo Category:")
category_length_stats = df.groupby('Category').agg({
    'Question_Length': ['mean', 'std', 'min', 'max'],
    'Answer_Length': ['mean', 'std', 'min', 'max'],
    'Question': 'count'
}).round(2)

print(category_length_stats)
````

### 3.2. Ph√¢n t√≠ch v√† x·ª≠ l√Ω d·ªØ li·ªáu tr√πng l·∫∑p


````python
# Ph√¢n t√≠ch d·ªØ li·ªáu tr√πng l·∫∑p
print("üîç PH√ÇN T√çCH D·ªÆ LI·ªÜU TR√ôNG L·∫∂P")
print("=" * 50)

# 1. Ki·ªÉm tra tr√πng l·∫∑p ho√†n to√†n
exact_duplicates = df.duplicated().sum()
print(f"üìä S·ªë d√≤ng tr√πng l·∫∑p ho√†n to√†n: {exact_duplicates}")

# 2. Ki·ªÉm tra tr√πng l·∫∑p theo c√¢u h·ªèi
question_duplicates = df.duplicated(subset=['Question']).sum()
print(f"üìä S·ªë c√¢u h·ªèi tr√πng l·∫∑p: {question_duplicates}")

# 3. Ki·ªÉm tra tr√πng l·∫∑p theo c√¢u h·ªèi v√† c√¢u tr·∫£ l·ªùi
qa_duplicates = df.duplicated(subset=['Question', 'Answer']).sum()
print(f"üìä S·ªë c·∫∑p Q&A tr√πng l·∫∑p: {qa_duplicates}")

# 4. Hi·ªÉn th·ªã c√°c c√¢u h·ªèi tr√πng l·∫∑p
print(f"\nüîç C√°c c√¢u h·ªèi tr√πng l·∫∑p:")
duplicate_questions = df[df.duplicated(subset=['Question'], keep=False)].sort_values('Question')
if len(duplicate_questions) > 0:
    print(f"T√¨m th·∫•y {len(duplicate_questions)} d√≤ng c√≥ c√¢u h·ªèi tr√πng l·∫∑p")
    
    # Nh√≥m theo c√¢u h·ªèi ƒë·ªÉ xem chi ti·∫øt
    duplicate_groups = duplicate_questions.groupby('Question').agg({
        'Topic': lambda x: ', '.join(x.unique()),
        'Category': lambda x: ', '.join(x.unique()),
        'Answer': lambda x: len(x.unique())
    }).reset_index()
    
    print(f"\nüìã Chi ti·∫øt c√°c c√¢u h·ªèi tr√πng l·∫∑p:")
    for i, (_, row) in enumerate(duplicate_groups.head(10).iterrows()):
        print(f"{i+1}. C√¢u h·ªèi: {row['Question'][:80]}...")
        print(f"   Ch·ªß ƒë·ªÅ: {row['Topic']}")
        print(f"   Category: {row['Category']}")
        print(f"   S·ªë c√¢u tr·∫£ l·ªùi kh√°c nhau: {row['Answer']}")
        print("-" * 60)
else:
    print("‚úÖ Kh√¥ng c√≥ c√¢u h·ªèi tr√πng l·∫∑p")

# 5. Ki·ªÉm tra tr√πng l·∫∑p g·∫ßn ƒë√∫ng (similarity)
from difflib import SequenceMatcher

def similarity(a, b):
    """T√≠nh ƒë·ªô t∆∞∆°ng ƒë·ªìng gi·ªØa 2 chu·ªói"""
    return SequenceMatcher(None, a.lower(), b.lower()).ratio()

print(f"\nüîç Ki·ªÉm tra tr√πng l·∫∑p g·∫ßn ƒë√∫ng (similarity > 0.8):")
similar_questions = []
questions_list = df['Question'].tolist()

# Ki·ªÉm tra t·ª´ng c·∫∑p c√¢u h·ªèi (ch·ªâ ki·ªÉm tra m·∫´u nh·ªè ƒë·ªÉ tr√°nh qu√° t·∫£i)
sample_size = min(1000, len(questions_list))
sample_questions = questions_list[:sample_size]

print(f"Ki·ªÉm tra {sample_size} c√¢u h·ªèi ƒë·∫ßu ti√™n...")

for i in range(len(sample_questions)):
    for j in range(i+1, len(sample_questions)):
        sim = similarity(sample_questions[i], sample_questions[j])
        if sim > 0.8:
            similar_questions.append({
                'question1': sample_questions[i],
                'question2': sample_questions[j],
                'similarity': sim
            })

if similar_questions:
    print(f"T√¨m th·∫•y {len(similar_questions)} c·∫∑p c√¢u h·ªèi t∆∞∆°ng ƒë·ªìng:")
    for i, sim_q in enumerate(similar_questions[:5]):  # Hi·ªÉn th·ªã 5 c·∫∑p ƒë·∫ßu
        print(f"{i+1}. ƒê·ªô t∆∞∆°ng ƒë·ªìng: {sim_q['similarity']:.3f}")
        print(f"   C√¢u 1: {sim_q['question1'][:60]}...")
        print(f"   C√¢u 2: {sim_q['question2'][:60]}...")
        print("-" * 60)
else:
    print("‚úÖ Kh√¥ng t√¨m th·∫•y c√¢u h·ªèi t∆∞∆°ng ƒë·ªìng")
````

````python
# Lo·∫°i b·ªè d·ªØ li·ªáu tr√πng l·∫∑p
print("\nüßπ LO·∫†I B·ªé D·ªÆ LI·ªÜU TR√ôNG L·∫∂P")
print("=" * 50)

# L∆∞u tr·ªØ d·ªØ li·ªáu g·ªëc ƒë·ªÉ so s√°nh
original_size = len(df)
print(f"üìä K√≠ch th∆∞·ªõc d·ªØ li·ªáu g·ªëc: {original_size:,} d√≤ng")

# 1. Lo·∫°i b·ªè tr√πng l·∫∑p ho√†n to√†n
df_no_exact_dup = df.drop_duplicates()
print(f"üìä Sau khi lo·∫°i b·ªè tr√πng l·∫∑p ho√†n to√†n: {len(df_no_exact_dup):,} d√≤ng")
print(f"   ƒê√£ lo·∫°i b·ªè: {original_size - len(df_no_exact_dup):,} d√≤ng")

# 2. Lo·∫°i b·ªè tr√πng l·∫∑p theo c√¢u h·ªèi (gi·ªØ l·∫°i d√≤ng ƒë·∫ßu ti√™n)
df_no_q_dup = df_no_exact_dup.drop_duplicates(subset=['Question'], keep='first')
print(f"üìä Sau khi lo·∫°i b·ªè tr√πng l·∫∑p c√¢u h·ªèi: {len(df_no_q_dup):,} d√≤ng")
print(f"   ƒê√£ lo·∫°i b·ªè th√™m: {len(df_no_exact_dup) - len(df_no_q_dup):,} d√≤ng")

# 3. Lo·∫°i b·ªè tr√πng l·∫∑p theo c·∫∑p Q&A
df_clean = df_no_q_dup.drop_duplicates(subset=['Question', 'Answer'], keep='first')
print(f"üìä Sau khi lo·∫°i b·ªè tr√πng l·∫∑p Q&A: {len(df_clean):,} d√≤ng")
print(f"   ƒê√£ lo·∫°i b·ªè th√™m: {len(df_no_q_dup) - len(df_clean):,} d√≤ng")

# T·ªïng k·∫øt
total_removed = original_size - len(df_clean)
print(f"\n‚úÖ T·ªîNG K·∫æT:")
print(f"   D·ªØ li·ªáu g·ªëc: {original_size:,} d√≤ng")
print(f"   D·ªØ li·ªáu sau khi l√†m s·∫°ch: {len(df_clean):,} d√≤ng")
print(f"   T·ªïng s·ªë d√≤ng ƒë√£ lo·∫°i b·ªè: {total_removed:,} d√≤ng ({total_removed/original_size*100:.1f}%)")

# C·∫≠p nh·∫≠t DataFrame ch√≠nh
df = df_clean.copy()

# Th·ªëng k√™ theo category sau khi l√†m s·∫°ch
print(f"\nüìä Th·ªëng k√™ theo category sau khi l√†m s·∫°ch:")
category_stats_clean = df['Category'].value_counts()
print(category_stats_clean)

# Hi·ªÉn th·ªã ph√¢n ph·ªëi ƒë·ªô d√†i sau khi l√†m s·∫°ch
print(f"\nüìä Th·ªëng k√™ ƒë·ªô d√†i sau khi l√†m s·∫°ch:")
print(f"C√¢u h·ªèi - Trung b√¨nh: {df['Question_Length'].mean():.1f} k√Ω t·ª±")
print(f"C√¢u h·ªèi - Trung v·ªã: {df['Question_Length'].median():.1f} k√Ω t·ª±")
print(f"C√¢u tr·∫£ l·ªùi - Trung b√¨nh: {df['Answer_Length'].mean():.1f} k√Ω t·ª±")
print(f"C√¢u tr·∫£ l·ªùi - Trung v·ªã: {df['Answer_Length'].median():.1f} k√Ω t·ª±")
````
